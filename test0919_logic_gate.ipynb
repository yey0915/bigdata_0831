```python
import numpy as np
import torch
import torch.nn as nn
torch.manual_seed(777)
```

```python
x = torch.FloatTensor([[0,0], [0,1], [1,0], [1,1]])
y = torch.FloatTensor([[0], [1], [1], [1]]) # XOR 는 값이 다르게 나옴
```

```python
model = nn.Sequential(
    nn.Linear(2,1),
    nn.Sigmoid()
)
```

```python
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for epoch in range(1001):
    h_hat = model(x)
    cost = criterion(h_hat, y)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print("Epoch : {}  Cost : {}".format(epoch, cost.item()))
```

```python
with torch.no_grad():
    pred = model(x)
    predicted = (pred > 0.5).float()
    print(pred)
    print(predicted)
```

- 다층 퍼셉트론

```python
x = torch.FloatTensor([[0,0], [0,1], [1,0], [1,1]])
y = torch.FloatTensor([[0], [1], [1], [0]]) # XOR 는 값이 다르게 나옴
```

```python
model = nn.Sequential(
    # Linear(2, 10) : 2개의 입력값과 은닉층 첫번째 노드의 크기가 10개라는 의미
    # (input_dim=2, output_dim=10)
    nn.Linear(2,10, bias=True), 
    nn.Sigmoid(), 
    nn.Linear(10,10, bias=True), # 은닉층 : 10 / (input_dim=10, output_dim=10)
    nn.Sigmoid(),
    nn.Linear(10, 10),          # 은닉층 : 10 / (input_dim=10, output_dim=10)
    nn.Sigmoid(),
    nn.Linear(10, 1, bias=True),    # 출력층 : 1 / (input_dim=10, output_dim=1)
    nn.Sigmoid()
)
```

```python
criterion1 = torch.nn.BCELoss() # BCELoss : 바이너리 크로스 손실함수
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
for epoch in range(100001):
    h_hat = model(x)
    cost = criterion1(h_hat, y)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()


    if epoch % 100 == 0:
        print("Epoch : {}  Cost : {}".format(epoch, cost.item()))
```

```python
with torch.no_grad():
    pred = model(x)
    predicted = (pred > 0.5).float()
    print(predicted)
```
