```python
titanic = sns.load_dataset('titanic')
print("타이타닉 전체 데이터 :\n", titanic)
```

```python
import math
def correlation(x, y):
    n = len(x)
    x_sum = 0.0
    y_sum = 0.0
    x_sum_pow = 0.0 # 제곱
    y_sum_pow = 0.0 # 제곱
    mul_xy_sum = 0.0

    for i in range(n) :
        mul_xy_sum = mul_xy_sum + float(x[i]) * float(y[i]) # xy 의 sum
        x_sum = x_sum + float(x[i])
        y_sum = y_sum + float(y[i])
        x_sum_pow = x_sum_pow + pow(float(x[i]), 2)
        y_sum_pow = y_sum_pow + pow(float(y[i]), 2)

        try:
            r = ((n*mul_xy_sum)-(x_sum*y_sum)) / math.sqrt(((n*x_sum_pow) - pow(x_sum, 2)) * ((n*y_sum_pow)-pow(y_sum, 2)))
        except :
            r = 0.0
        
    return r 
```

```python
print("titanic의 상관관계 :\n", titanic.corr())
```

```python
print('survived와 pclass에 대한 상관관계 : ', correlation(titanic['survived'], titanic['pclass']))
print('survived와 sibsp 사이의 상관관계 : ', correlation(titanic['survived'], titanic['sibsp']))
print('survived와 adult_male 사이의 상관관계 : ', correlation(titanic['survived'], titanic['adult_male']))
```
