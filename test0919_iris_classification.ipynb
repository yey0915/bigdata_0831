```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from sklearn import datasets

iris = datasets.load_iris()
print(iris.data)
print(iris.target)
print(iris.data.shape)
print(iris.target.shape)
```

```python
x_train = torch.FloatTensor(iris.data)
y_train = torch.LongTensor(iris.target)
torch.manual_seed(1)
```

- python 기본 소스로 모델 작성
- 클래스로 모델 작성

```python
y = torch.randint(5, (150,)).long()
print(y)
```

```python
y_one_hat = torch.zeros(150,3)
print(y_one_hat)
```

```python
print(y_train.unsqueeze(1))
y_one_hat.scatter_(1, y_train.unsqueeze(1), 1)
```

```python
print(y_train[:5])
print(y_one_hat[:5])
```

```python
w = torch.zeros((4, 3), requires_grad=True) # 4, 3 : 3은 3가지의 방법을 의미
b = torch.zeros((1, 3), requires_grad=True)
```

```python
pred = F.softmax(x_train.matmul(w)+b, dim=1).argmax(dim=1)
acuracy = (pred == y_train).sum() / len(pred)
print(acuracy)
```

```python
optimizer = optim.SGD([w, b], lr=0.1)
```

```python
for epoch in range(10001):
    y_hat = F.softmax(x_train.matmul(w)+b, dim=1)
    cost = (y_one_hat * -torch.log(y_hat)).sum(dim=1).mean()

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 400 == 0:
        print("Epoch : {} Cost : {}".format(epoch, cost.item()))
```

```python
class SoftmaxClassModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(4,3)

    def forward(self, x):
        return self.linear(x)
```

```python
model = SoftmaxClassModel()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
for epoch in range(10001):
    y_hat = model(x_train)
    cost = F.cross_entropy(y_hat, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 400 == 0:
        accuracy = (y_hat.argmax(dim=1) == y_train).sum()/len(y_hat)
        print("Epoch : {} Cost: {}  Accuracy : {}".format(epoch, cost.item(), accuracy))
```
