```python
from nltk import Text
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
fontname='c:/Windows/fonts/HMFMMUEX.TTC'
font_name=font_manager.FontProperties(fname=fontname).get_name()
rc('font',family=font_name)
plt.figure(figsize=(6,4))
```

```python
okt.nouns(c) # 헌법에서 맹사 추출
kolaw_text = Text(okt.nouns(c))
kolaw_text.plot(30)
plt.show()
```

```python
# 빈도수
kv = kolaw_text.vocab()
kv
```

```python
# 빈도수 많은 단어 중에서 한자짜리 글자 제거 / 글자가 2자 이상인 것만 선택
kv2 = dict()  # 딕셔너리로 만듬
for tag, count in kv.items() :
    if len(tag) > 1 :
        kv2[tag] = count

kv2
```

```python
from wordcloud import WordCloud
font_path = 'c:/Windows/fonts/HMFMMUEX.TTC'
wc = WordCloud(width=1000, height=800, background_color='white', random_state=0, font_path=font_path)
plt.imshow(wc.generate_from_frequencies(kv2))
plt.axis('off')
plt.savefig('data/law.png')
plt.show()
```

```python
# Counter : 문자열 안에 단어별로 갯수 확인
from collections import Counter
s2 = 'aaaabbbcc'
s3 = '한글도 가능한가 궁금하다'
sarr = list(s2)
sarr_kr = list(s3)
# print(sarr)
# Counter(sarr)
sarr_kr
Counter(sarr_kr)
```

```python
nouns1 = okt.nouns(c)
counter1 = Counter(nouns1)
# print(counter1)
counter2 = counter1.most_common(10) # most_common(10) : 빈도수 많은것 10개 추출
print(counter2)

counter3 = dict()
for tag, count in counter1.items() :
    if (len(tag) > 1) and (count>1) : # 글자수 2자 이상, 빈도수 2이상인 글자만 추출
        counter3[tag] = count 
counter3
```

```python
wc1 = WordCloud(width=800, height=600, font_path=font_path, background_color='white')
plt.imshow(wc1.generate_from_frequencies(counter1))
plt.axis('off')
plt.show()
```

```python
data = open('data/data.txt').read()
data
```

- 명사추출
- 형태소 추출
- 품사 추출

```python
data_noun = okt.nouns(data)
data_count = Counter(data_noun)

data_count2 = dict()
for tag, count in data_count.items() :
    if (len(tag) > 1) and (count > 1) :
        data_count2[tag] = count

data_count2
```

```python
import numpy as np
from PIL import Image

im = Image.open('data/leaf.png')
mask_arr = np.array(im) # WordCloud : 모양 지정
mask_arr
```

```python
wc2 = WordCloud(width=400, height=400, background_color='white', font_path=font_path, mask=mask_arr)
plt.imshow(wc2.generate_from_frequencies(data_count2))
plt.axis('off')
plt.show()
```
