```python
import numpy as np
import pandas as pd
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
raw_df
```

```python
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2,  :2]]) # 리스트로 들어옴
data
```

```python
target = raw_df.values[1::2, 2]
target
```

```python
df = pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
df.head()
```

```python
df['TARGET'] = target
df.head()
```

요약통계
1. chas에 따른 집가격(target)의 평균, 표준편차, 중앙값을 구하라.
2. CHAS(찰스강 경계) 따른 집값에 대한 t테스트 수행
3. 선형회귀분석을 모델을 생성하고 모델을 사용하여 예측값을 구하라.
4. 새로운 데이터 2개를 사용하여 예측값을 구하라.
5. 부분회귀분석 시각화를 작성하여 독립변수들이 종속변수에 미치는 영향을 분석하라. 
 - 1개독립변수 분석(plot_partregress)은 (CRIM)으로한다.
 - plot_partregress_grid() 작성


```python
# 1. chas에 따른 집가격(target)의 평균, 표준편차, 중앙값을 구하라.
df.groupby('CHAS')['TARGET'].agg(['mean', 'std', 'median'])
```

```python
# 2. CHAS(찰스강 경계)에 따른 집값에 대한 t테스트 수행
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
chas1 = df.loc[df["CHAS"] == 0, 'TARGET']
chas0 = df.loc[df["CHAS"] == 1, 'TARGET']
```

```python
stats.ttest_ind(chas1, chas0, equal_var=False)
```

```python
# 3. 선형회귀분석을 모델을 생성하고 모델을 사용하여 예측값을 구하라.
Rformula = "TARGET ~ CRIM + ZN + INDUS +  CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT"
```

```python
regression_result = ols(Rformula, data=df).fit() # ols 로 훈련을 시킨다
regression_result
```

```python
# 4. 새로운 데이터 2개를 사용하여 예측값을 구하라.
data = {
    "CRIM" : [0.03245, 0.11876], 
    "ZN":[60, 93],
    "INDUS":[3.25, 6.12], 
    "CHAS":[0, 1], 
    "NOX":[0.521, 0.321], 
    "RM":[6.255, 4.125], 
    "AGE":[36, 68.2], 
    "DIS":[3.09, 6.123], 
    "RAD":[3, 9], 
    "TAX":[254, 396],
    "PTRATIO":[14.3, 17],
    "B":[396.9, 365.8],
    "LSTAT":[18.8, 9.96]}
```

```python
sample1 = df[df.columns.difference(['TARGET'])]
sample1.head()
```

```python
sample2 = pd.DataFrame(data, columns=sample1.columns)
sample2
```

```python
sample2_predict = regression_result.predict(sample2)
sample2_predict
```
