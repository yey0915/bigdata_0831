```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
import numpy as np
import pandas as pd
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df=pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
raw_df
```

```python
torch.manual_seed(1)
data=np.hstack([raw_df.values[::2, :], raw_df.values[1::2,:2]])
x_data = torch.FloatTensor(data)
print(x_data)
print(x_data.size())
# x 값 받기
ax = x_data.size()[1]
```

```python
target=raw_df.values[1::2, 2]
y_data = torch.FloatTensor(target)
y_data = y_data.view(506,1)
print(y_data)
print(y_data.size())
#y 값 받기
ay = y_data.size()[1]
```

```python
class MultiLinearRegressionModel(nn.Module):
    # 생성자 생성 : 생성자에서는 모델을 정의한다.
    def __init__(self, ax, ay):
        super().__init__()
        self.linear = nn.Linear(ax,ay)
    
    # 정의한 모델 실행
    def forward(self, x):
        return self.linear(x)
```

```python
model = MultiLinearRegressionModel(ax, ay)
optimizer = optim.SGD(model.parameters(), lr=1e-7)
```

```python
for epoch in range(200001):
    prediction = model(x_data)
    cost = F.mse_loss(prediction, y_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 400 == 0:
        print('Epoch: ', epoch, 'cost :', cost.item())
```
