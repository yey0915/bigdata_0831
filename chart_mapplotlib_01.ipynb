```python
# 2차원 데이터 만들기
# data4 = np.random.rand(10, 4)
# print(data4)
data4 = np.random.randint(1, 100, size=(10, 4)) #정수로 만들고 싶을때 : size <- 차원의 크기를 정할 수 있음
# print(data4)
df = pd.DataFrame(data4, columns=['A', 'B', 'C', 'D'])
print(df)
```

```python
# df.plot(kind='bar')
# df.plot(kind='barh')
df.plot(kind='barh', stacked=True) #kind=bar 일 경우 stacked의 기본값은 False
```

```python
x1 = np.random.normal(1, 1, size=(100, 1))
x2 = np.random.normal(-2, 4, size=(100, 1))
x = np.concatenate((x1, x2), axis=1) # numpy 배열 두개를 합친다. / axis = 1, 0 은 두개의 배열을 행으로 붙일것인가 열로 붙일것인가를 결정
# x = np.concatenate((x1, x2), axis=0)
x
df1 = pd.DataFrame(x, columns=['x1', 'x2'])
df1
```

```python
filepath = 'data/temperature_2014.csv'
df = pd.read_csv(filepath)
# df.info() # 데이터 프레임은 꼭 형을 확인해야 한다
df['PST'] = pd.to_datetime(df['PST'])
# df.info()
fig = plt.figure(figsize=(10, 6))
plt.plot(df['PST'], df['Max TemperatureF'], c='red', alpha=0.5)
plt.plot(df['PST'], df['Min TemperatureF'], c='blue', alpha=0.5)
plt.fill_between(df['PST'], df['Max TemperatureF'], df['Min TemperatureF'], facecolor='#CAEDFF')
plt.title('온도 통계', fontsize=20)
plt.xlabel('기간', fontsize=16)
plt.ylabel('온도', fontsize=16)
plt.tick_params(axis='both', labelsize=16)
plt.savefig('data/chart.png')
plt.show()
```

```python
filepath = 'data/temperature_2014.csv'
df = pd.read_csv(filepath)
df['PST'] = pd.to_datetime(df['PST'])
df.info()
plt.figure(figsize=(6, 10))
plt.subplot(3,1,1)
plt.plot(df['PST'], df['Min TemperatureF'])
plt.title('최저기온', fontsize='12')
plt.subplot(3,1,2)
plt.plot(df['PST'], df['Mean TemperatureF'])
plt.title('평균기온', fontsize='12')
plt.subplot(3,1,3)
plt.plot(df['PST'], df['Max TemperatureF'])
plt.title('최고기온', fontsize='12')
plt.show()
```
