```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rc, font_manager
```

```python
sns.set(style='whitegrid') 

fontpath = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=fontpath).get_name() 
rc('font', family=font_name)

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12,9)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# sns.catplot(x='day', y='total_bill', data=tips)
# sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
# sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```

```python
# violin 챠트는 카테고리 별로 분류를 할때 어느 구간에 몰려 있는지 확인 할때 사용
# sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
# sns.catplot(x='day', y='total_bill', data=tips, kind='violin', inner='stick')
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex')
```

```python
# 많이 사용함
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True)
```

```python
# inner='stick' : 실제 값이 분포 되여 있는 곳
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='smoker', split=True, inner='stick')
```
