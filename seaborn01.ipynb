```python
# key=time / value = 1~500 사이값
# cumsum : 누적의 합
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum())) 
# 아래 두개는 같은 결과를 낸다.
sns.relplot(x='time', y='value', data=df, kind='line')
sns.lineplot(x='time', y='value', data=df)
```

seaborn - heatmap : 데이터 값을 색상으로 강조하여 표현

```python
data = np.random.rand(5, 6) # 5행 6열의 데이터 생성
print(data)
sns.heatmap(data, annot=True)
```

```python
corr = tips.corr() # 숫자만으로 상관관계를 설정
# corr
sns.heatmap(corr, annot=True) # annot=True : 어노테인션을 표기, 네모안에 값을 annot라고 한다.
```

```python
sns.heatmap(corr, annot=True, cmap='GnBu')
```
