# 상관관곕문석과 상관관계
- 상관관계분석 : 두 변수 사이의 상관정도를 분석
- 상관계수 : 상관관계 강도를 나타내는 것
- 상관관계 분석식 :
- 표본

```python
import math
def correlation(x, y):
    n = len(x)
    x_sum = 0.0
    y_sum = 0.0
    x_sum_pow = 0.0 # 제곱
    y_sum_pow = 0.0 # 제곱
    mul_xy_sum = 0.0

    for i in range(n) :
        mul_xy_sum = mul_xy_sum + float(x[i]) * float(y[i]) # xy 의 sum
        x_sum = x_sum + float(x[i])
        y_sum = y_sum + float(y[i])
        x_sum_pow = x_sum_pow + pow(float(x[i]), 2)
        y_sum_pow = y_sum_pow + pow(float(y[i]), 2)

        try:
            r = ((n*mul_xy_sum)-(x_sum*y_sum)) / math.sqrt(((n*x_sum_pow) - pow(x_sum, 2)) * ((n*y_sum_pow)-pow(y_sum, 2)))
        except :
            r = 0.0
        
    return r    
```

```python
print(f"키 : \n{height}")
print(f"발사이즈 : \n{foot_size}")
print('키와 발크기에 대한 상관관계 : ', correlation(height, foot_size))
print('산높이와 온도 사이의 상관관계 : ', correlation(h, t))
print('임의의 x1에 대한 x2의 상관관계 : ', correlation(x1, x2))
```

```python
# 위의 자료로 heatmap을 만들어보자
# annot=True : 데이터를 보여주는 설정
# annot_kws={'size':12} : 글자사이즈
# fmt='.3f' : 소숫점 3자리까지 표시
# square=True : 정사각형으로 만들지에 대한 설정
# cmap='Reds' : 색깔 설정
df_corr_heatmap = sns.heatmap(df_corr, cbar=True, annot=True, annot_kws={'size':12}, fmt='.3f', square=True, cmap='Blues', vmin=0.1, vmax=1 )
# sns.color_palette("light:#5A9", as_cmap=True)
```
