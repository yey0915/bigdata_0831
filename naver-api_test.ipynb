```python
from nltk import Text
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from collections import Counter
```

```python
from konlpy.tag import *

okt = Okt()
okt.nouns(str_title)
```

```python
# 뉴스
title_nouns = okt.nouns(str_title)
title_nouns
title_count = Counter(title_nouns)
title_count
```

```python
counter = dict()
for tag, count in title_count.items() :
    if (len(tag) > 1) and (count >1) :
        counter[tag] = count
counter
```
