```python
# 네이버 검색 API 예제 - 블로그 검색
import os # 경로 / 파일 사용시 필요
import sys
import json
import urllib.request # 웹에 있는 데이터를 가지고 옴

client_id = 'X2EgnFYeuveMTgHqwr7X'
client_secret = 'Zothq1xjvX'
encText = urllib.parse.quote("다이어트") # utf-8로 인코딩
url = "https://openapi.naver.com/v1/search/news?query=" + encText # JSON 결과 / 네이버 뉴스 가지고 옴
# url = "https://openapi.naver.com/v1/search/news?query=" + encText + "&display=100" # 뉴스 100개 가지고 옴
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url) # request를 하나 만듬
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret", client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode() # 200 / 404 / 500 등등......

str_title = '' # 명사 추출 / 명사 빈도수 구하기 (단어 짧은것 / 글자수 작은것 제외)
if(rescode==200):
    response_body = response.read()
    res_body = json.loads(response_body.decode('utf-8'))
    for item in res_body['items'] :
        # print(item)
        # print(item['title'])
        str_title += item['description'] + ' '
        print(item['description'])
    # print(res_body['items'])
else:
    print("Error Code:" + rescode)

```
