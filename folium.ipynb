```python
import folium
```

- location : 지도의 중심 좌표(위도, 경도)
- zoom_start : 지도의 확대 정도(1 ~ 18, default = 10)
- tiles : 지도의 지형 모양 ( 예 : 산악지형 등등 ) : Stamen Terrain, Open StreetMap, Stamen Toner

```python
latitude = 35.1795543
longitude = 129.0756416
```

```python
# 보여지는 지도의 중앙으로 표시 : location / zoom_start : 1 ~ 20 사이의 값 선택
m = folium.Map(location=[latitude, longitude], zoom_start=20) 
# m = folium.Map(location=[latitude,longitude], zoom_start=12, tiles='Stamen Terrain') # 산악지형
# m = folium.Map(location=[latitude,longitude], zoom_start=12, tiles='Open StreetMap')  # 특별히 정하지 않은 경우
# m = folium.Map(location=[latitude,longitude], zoom_start=12, tiles='Stamen Toner')
# m.save('data/save/map1.html')
folium.Marker([latitude, longitude], popup='부산시청').add_to(m)
m
```

```python
#덕수궁
latitude1 = 37.56583
longitude1 = 126.97472
```

```python
m = folium.Map(location=[latitude1, longitude1], zoom_start=20) 
folium.Marker([latitude1, longitude1], popup='덕수궁').add_to(m)
m
```

```python
m = folium.Map(location=[latitude1, longitude1], zoom_start=20) 
folium.Marker([latitude1, longitude1], popup='덕수궁', icon=folium.Icon(color='red', icon_color='yellow', icon='info-sign')).add_to(m)
m
```
